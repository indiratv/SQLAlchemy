```python
# Import all dependecies
import datetime as dt
import time
import numpy as np
import pandas as pd

from sqlalchemy import create_engine, inspect
import sqlalchemy 
from sqlalchemy.ext.automap import automap_base
# pymysql.install_as_MySQLdb()

from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float, DateTime,Date
from sqlalchemy import and_
from sqlalchemy.orm import sessionmaker
from sqlalchemy.orm import Session
from flask import Flask, jsonify
from sqlalchemy import func

import matplotlib
from matplotlib import style
style.use('seaborn')
import matplotlib.pyplot as plt
```

```python
#Connect to hawaii sqllite database
engine = create_engine("sqlite:///hawaii.sqlite")
conn = engine.connect()
#Create a session for this connecion
session = Session(bind=conn)
```

```python
#Reflect the tables as Classes
Base = automap_base()
```

```python
Base.prepare(engine, reflect=True)
```

```python
#Get all the classes corresponding to the tables in the database
Base.classes.keys()
```

```python
# Create instances of the tables reflected as classes
Station = Base.classes.Station
Measurement = Base.classes.Measurement
```

```python
# Create a session
session = Session(engine)
```

```python
# Create the inspector and connect it to the engine
inspector = inspect(engine)
```

```python
#Describe the columns for the Measurement class
col = inspector.get_columns('Measurement')
for column in col:
    print(column["name"], column["type"])
```

```python
#Precipitation Analysis
#Import dependencies for date and time calculations

import datetime as dt
from dateutil.relativedelta import relativedelta
from sqlalchemy import and_
```

```python
#Calculate today and 1 yr from today in the past
today = dt.date.today()
today = dt.date(day=today.day, month=today.month, year=today.year)
today1yrback = today - relativedelta(days=365)
today1yrback
```

```python
#Get Precipitation results in the last one year
measurement_results = session.query(Measurement.date, Measurement.prcp).\
    filter(and_(Measurement.date>=today1yrback,Measurement.date<=today)).all()
measurement_results
```

```python
measurement_results_date = session.query(Measurement.date).\
    filter(and_(Measurement.date>=today1yrback,Measurement.date<=today)).all()
measurement_results_date
```

```python
datel = [date[0] for date in measurement_results_date]
datel
```

```python
#Create a dataframe of Precipitation results in the last one year and set the date as index
measurement_results_df = pd.DataFrame(measurement_results, columns=['date', 'prcp'],index=(datel))
measurement_results_df.head()
```

```python
# Plot Precipitation during the last 12 months
plt.figure(figsize=(20,10))
plt.grid(b=True, linestyle='-')  
plt.xlabel("Day of the Year")
plt.ylabel("Precipitation")
plt.title("Precipitation during last 12 months")
plt.bar(measurement_results_df["date"],measurement_results_df["prcp"],color='skyblue',label= 'Precipitation',lw=10,alpha=1)
plt.locator_params(axis='x', nbins=12,tight=False)
plt.show()
```

```python
#Temperature Analysis
#Function to calculate minimum, average, and maximum temperatures for a given range of dates
import datetime as dt
def calc_temp(start_date,end_date):
    sd = dt.datetime.strptime(start_date, '%Y-%m-%d')
    ed = dt.datetime.strptime(end_date, '%Y-%m-%d')
    startdate1yrback = sd - relativedelta(days=365)
    endat1yrback = ed - relativedelta(days=365)
    results = session.query(func.min(Measurement.tobs).label("mintemp"),func.max(Measurement.tobs).label("maxtemp"),func.avg(Measurement.tobs).label("avgtemp")).\
    filter(and_(Measurement.date>=startdate1yrback,Measurement.date<= endat1yrback)).all()
    return results;
```

```python
#Enter start_date and end_date for temperature querying from dataset for the previous year
start_date = input('Enter Start date in the format y-m-d: ' '%Y-%m-%d')
end_date =   input('Enter End date in the format y-m-d: ' '%Y-%m-%d')
```

```python
# Call teh function to show min,max and avg temp.
resu = calc_temp(start_date,end_date)
mintemp = resu[0][0]
maxtemp = resu[0][1]
avgtemp = resu[0][2]
```
